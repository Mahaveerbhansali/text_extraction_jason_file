```python
import os
import pdfplumber
from bs4 import BeautifulSoup

def extract_content(file_path):
    # Check file existence
    if not os.path.exists(file_path):
        return f"Error: File does not exist at {file_path}"

    # Determine the file type
    file_extension = os.path.splitext(file_path)[1].lower()
    
    if file_extension == ".pdf":
        return extract_from_pdf(file_path)
    elif file_extension in [".html", ".htm"]:
        return extract_from_html(file_path)
    else:
        return f"Error: Unsupported file format {file_extension}"

def extract_from_pdf(file_path):
    try:
        with pdfplumber.open(file_path) as pdf:
            content = ""
            for page in pdf.pages:
                content += page.extract_text()
        return content if content.strip() else "Error: No text found in PDF"
    except Exception as e:
        return f"Error while processing PDF: {str(e)}"

def extract_from_html(file_path):
    try:
        # Open and parse the HTML file
        with open(file_path, "r", encoding="utf-8") as file:
            soup = BeautifulSoup(file, "lxml")
        
        # Extract all text while keeping the structure clean
        # Optional: Add more specific parsing logic if needed
        content = soup.get_text(separator="\n", strip=True)
        
        # Return the text if available
        return content if content.strip() else "Error: No text found in HTML"
    except Exception as e:
        return f"Error while processing HTML: {str(e)}"


# File paths
pdf_path = "C:/Users/mahav/OneDrive - M S Ramaiah University of Applied Sciences/Desktop/emplay/Bid1/PORFP_-_Dell_Laptop_Final.pdf"
html_path = "C:/Users/mahav/OneDrive - M S Ramaiah University of Applied Sciences/Desktop/emplay/Bid2/Dell Laptops w_Extended Warranty - Bid Information - {3} _ BidNet Direct.html"

# Extract and print content
pdf_content = extract_content(pdf_path)
html_content = extract_content(html_path)

print("PDF Content:")
print(pdf_content[:1000])  # Preview the first 1000 characters

print("\nHTML Content:")
print(html_content[:1000])  # Preview the first 1000 characters
```

```python
import os
import re
import pdfplumber
import json

# Function to extract content from the PDF
def extract_from_pdf(file_path):
    try:
        with pdfplumber.open(file_path) as pdf:
            content = ""
            for page in pdf.pages:
                content += page.extract_text()
        return content if content.strip() else "Error: No text found in PDF"
    except Exception as e:
        return f"Error while processing PDF: {str(e)}"

# Define regex patterns for key fields
patterns = {
    "PORFP Number": r"PORFP Number:\s*(\S+)",
    "eMMA Project Number": r"eMMA Project Number:\s*(\S+)",
    "PORFP Type": r"PORFP Type:\s*(\S+)",
    "Functional Area/s": r"Functional Area/s \(FA\) for this PORFP:\s*(.*?)(?=\n|$)",
    "Manufacturer Name": r"Manufacturer Name\s*([A-Za-z\s]+)",
    "Small Business Reserve": r"Designated Small Business Reserve\?\(SBR\):\s*(\S+)",
    "Minority Business Enterprise (MBE) Goal": r"Minority Business Enterprise \(MBE\) Goal.*?(\d+)%",  
    "Place of Performance": r"Place of Performance:\s*([A-Za-z0-9\s,]+)",
    "Proposal Due Date": r"PROPOSAL DUE DATE and TIME:\s*([\d/]+)",
    "Bid Submission Instructions": r"BID SUBMISSION INSTRUCTIONS\s*(.*?)\n",  # Matches the bid submission instructions section
    "Security Requirements": r"Security Requirements.*?(\d\..+?)\n",  # Matches the security requirements section
    "Evaluation Criteria": r"Evaluation Criteria.*?(\d\..+?)\n",  # Matches evaluation criteria
    "Agency POC Name": r"Agency POC Name:\s*([A-Za-z\s]+)",
    "Agency POC Email": r"Agency POC Email Address:\s*([\w\.\-]+@[\w\.\-]+)",
    "Product Specifications": r"Product Name\s*Product Description\s*Model #\s*Qty\s*(.*?)\s*Due Date",
    "Questions Due Date": r"Questions Due \(Closing\) Date and Time:\s*([\d/]+)",
    "Place of Performance": r"Place of Performance:\s*([\w\s,]+)"
}

# Function to extract fields using regex patterns
def extract_fields(text, patterns):
    extracted_data = {}
    
    for field, pattern in patterns.items():
        match = re.search(pattern, text, re.DOTALL)  # Allow multi-line matching
        if match:
            extracted_data[field] = match.group(1).strip()
        else:
            extracted_data[field] = "Not Found"
    
    return extracted_data

# Function to process PDF and extract fields
def process_pdf(pdf_path):
    # Extract PDF content
    pdf_content = extract_from_pdf(pdf_path)

    if "Error" in pdf_content:
        print(pdf_content)  # If there's an error with the PDF extraction
        return

    # Extract fields from the content using regex
    extracted_data = extract_fields(pdf_content, patterns)

    # Print the extracted data in JSON format
    print(json.dumps(extracted_data, indent=2))

# File path to the PDF
pdf_path = "C:/Users/mahav/OneDrive - M S Ramaiah University of Applied Sciences/Desktop/emplay/Bid1/PORFP_-_Dell_Laptop_Final.pdf"

# Process the PDF
process_pdf(pdf_path)
```

```python
import re
import json
from bs4 import BeautifulSoup

# Define patterns for key fields
patterns_html = {
    "PORFP Number": r"(PORFP Number):\s*(\S+)",
    "eMMA Project Number": r"(eMMA Project Number):\s*(\S+)",
    "PORFP Type": r"(PORFP Type):\s*(\S+)",
    "Functional Area/s": r"(Functional Area/s \(FA\) for this PORFP):\s*(.*?)(?=\n|$)",
    "Manufacturer Name": r"(Manufacturer Name):\s*([A-Za-z\s]+)",
    "Small Business Reserve": r"(Designated Small Business Reserve\?\(SBR\)):\s*(\S+)",
    "Minority Business Enterprise (MBE) Goal": r"(Minority Business Enterprise \(MBE\) Goal).*?(\d+)%",  
    "Proposal Due Date": r"(PROPOSAL DUE DATE and TIME):\s*([\d/]+)",
    "Bid Submission Instructions": r"(BID SUBMISSION INSTRUCTIONS)\s*(.*?)\n",
    "Security Requirements": r"(Security Requirements).*?(\d\..+?)\n",
    "Evaluation Criteria": r"(Evaluation Criteria).*?(\d\..+?)\n",
    "Agency POC Name": r"(Agency POC Name):\s*([A-Za-z\s]+)",
    "Agency POC Email": r"(Agency POC Email Address):\s*([\w\.\-]+@[\w\.\-]+)",
    "Product Specifications": r"(Product Name\s*Product Description\s*Model #\s*Qty)\s*(.*?)\s*Due Date",
    "Questions Due Date": r"(Questions Due \(Closing\) Date and Time):\s*([\d/]+)"
}

# Function to clean and process the raw text
def clean_and_process_text(raw_text):
    # Remove redundant whitespace and line breaks
    cleaned_text = re.sub(r"\s+", " ", raw_text)
    
    # Remove unnecessary headers, footers, or unrelated sections
    start_keywords = "BPM044557"  # Example keyword to identify where relevant text starts
    end_keywords = "This solicitation is CLOSED"  # Example end point for relevant text
    start_index = cleaned_text.find(start_keywords)
    end_index = cleaned_text.find(end_keywords)
    relevant_text = cleaned_text[start_index:end_index] if start_index != -1 and end_index != -1 else cleaned_text
    
    return relevant_text

# Function to extract fields using regex
def extract_fields(text, patterns):
    extracted_data = {}
    for field, pattern in patterns.items():
        match = re.search(pattern, text, re.DOTALL)
        extracted_data[field] = match.group(2).strip() if match else "Not Found"
    return extracted_data

# Function to process the HTML file and save data as JSON
def process_html_to_json(html_path, output_json_path):
    try:
        # Load the HTML file
        with open(html_path, "r", encoding="utf-8") as file:
            soup = BeautifulSoup(file, "html.parser")
        
        # Extract raw text from the HTML
        raw_text = soup.get_text(separator="\n", strip=True)
        
        # Clean and process the raw text
        processed_text = clean_and_process_text(raw_text)
        
        # Extract fields from the processed text
        extracted_data = extract_fields(processed_text, patterns_html)
        
        # Save extracted data to JSON
        with open(output_json_path, "w", encoding="utf-8") as json_file:
            json.dump(extracted_data, json_file, indent=2)
        
        print(f"Data successfully extracted and saved to {output_json_path}")
    
    except Exception as e:
        print(f"Error processing HTML file: {str(e)}")

# File paths
html_path = "C:/Users/mahav/OneDrive - M S Ramaiah University of Applied Sciences/Desktop/emplay/Bid2/Dell Laptops w_Extended Warranty - Bid Information - {3} _ BidNet Direct.html"
output_json_path = "C:/Users/mahav/OneDrive - M S Ramaiah University of Applied Sciences/Desktop/emplay/extracted_html_data.json"

# Process the HTML and save to JSON
process_html_to_json(html_path, output_json_path)
```
